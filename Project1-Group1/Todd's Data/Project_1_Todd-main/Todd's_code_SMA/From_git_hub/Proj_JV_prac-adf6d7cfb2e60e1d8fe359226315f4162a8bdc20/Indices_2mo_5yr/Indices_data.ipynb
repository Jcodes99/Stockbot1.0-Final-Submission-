```python
#pip install yahoo-finance

import os
import numpy as np 
import pandas as pd 
import yfinance as yf
import requests
import alpaca_trade_api as tradeapi
```

```python
# retrieving yahoo indices for cryptocurrency
crypto_indices = pd.read_html(requests.get('https://finance.yahoo.com/cryptocurrencies').text)
major_crypto = crypto_indices[0]
major_crypto.head()
```

```python
#Collecting BTC-USD
btc_data = yf.Ticker('BTC-USD')
```

```python
#Creating BTC history data for 2 months
btc_df = btc_data.history(period='2mo', start='2022-02-05', end='2022-05-05')
```

```python
# View header and tail of BTC for 2 months
print("BTC-USD data frame from: Feb 05, 20220 - May 05, 2022")
print("-----------------------------------------------------")
display(btc_df.head())
display(btc_df.tail())
```

```python
# Dropping last 2 columns of BTC 
btc_new_df = btc_df.drop(columns=['Stock Splits','Dividends'], axis=1)
print("BTC-USD update:")
print("---------------")
btc_new_df
```

```python
#csv BTC-USD file
btc_new_df.to_csv('BTC-USD.csv',float_format='%.2f') 
```

```python
# Retrieve 5 years history for BTC
btc_df = btc_data.history(period='5yr', start='2018-02-05', end='2022-05-05')
```

```python
# View header and tail of BTC for 5 years
print("BTC-USD data frame from: Feb 05, 2018 - May 05, 2022")
print("-----------------------------------------------------")
display(btc_df.head())
display(btc_df.tail())
```

```python
# Dropping last 2 columns of BTC 
btc_new_df = btc_df.drop(columns=['Stock Splits','Dividends'], axis=1)
print("BTC-USD update 2018-2022:")
print("-------------------------")
btc_new_df
```

```python
btc_new_df.to_csv('BTC-USD_5yr_his.csv',float_format='%.2f') 
```
