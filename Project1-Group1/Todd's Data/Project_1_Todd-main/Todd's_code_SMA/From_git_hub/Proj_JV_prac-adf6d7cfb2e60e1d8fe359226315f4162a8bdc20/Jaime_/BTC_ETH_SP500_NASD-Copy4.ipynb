```python
"""THIS BLOCK OF CODE DEFINES THE FUNCTION BTC_SMA_50() THAT CREATES THE DATAFRAME OF THE SIMPLE 50 DAY MOVING AVERAGE USING THE USER SPECIFIED RANGE OF TIME"""

def BTC_SMA_50():
    n = 50
    n_window = n + window
    btc_avg = btc_new_df
    
    def set_timeperiod():
        from datetime import date
        today = (date.today())
        print(today)

        from datetime import datetime, timedelta
        import time
        today = date.today()
        n = get_user_info_1()
        start_date = (date.today() - timedelta(days=n))

        return today, start_date

    while n in range(set_timeperiod(start_date),set_timeperiod(today)):
        #valid==(n < 60)
        n_window = n + window
        time = btc_new_df[n:n_window].sum()
        avg = time/window
        btc_avg = btc_avg.append(avg, ignore_index=False)
        n += 1 

    
    btc__50DMA_final = btc_avg.drop_duplicates()
    print("This is the listing of 50 day moving averages for Bitcoin")
    display(btc__50DMA_final)
    
    return btc__50DMA_final
```
